# Pie plot for species of test split

```python
import os
import plotly.express as px
import pandas as pd
```

```python
path = "/nfs/scratch/pinder/negative_dataset/my_repository/datasets/no_duplicates/deleak_cdhit/fully_balanced/test_species_counts.tsv"

df = pd.read_csv(path, sep="\t")

# print first few rows
print(df.head())
```

```python
# pie chart using plotly
fig = px.pie(df, names="Tax ID", values="Count in Test Set", title="Test Species Distribution")
fig.update_traces(textinfo="none")
fig.update_layout(width=1600, height=800)
fig.show()
```
