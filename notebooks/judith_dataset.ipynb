# Sequences in negative but not in positive part of a split? 

```python
judith_path = "/nfs/scratch/pinder/negative_dataset/my_repository/datasets/judith_gold_standard"
judith_train = pd.read_csv(os.path.join(judith_path, "train.csv"))
judith_val = pd.read_csv(os.path.join(judith_path, "val.csv"))
judith_test = pd.read_csv(os.path.join(judith_path, "test.csv"))
```

```python
import pandas as pd

# Sequences (either receptor or ligand) that appear only in label 0 pairs (never in any label 1 pair)

def sequences_only_in_label0(df, receptor_col="receptor_seq", ligand_col="ligand_seq", label_col="label"):
    label1_seqs = set(df.loc[df[label_col] == 1, receptor_col]) | set(df.loc[df[label_col] == 1, ligand_col])
    label0_receptor = set(df.loc[df[label_col] == 0, receptor_col])
    label0_ligand = set(df.loc[df[label_col] == 0, ligand_col])
    label0_all = label0_receptor | label0_ligand
    only_label0 = label0_all - label1_seqs

    # Determine in which role(s) they occur (within label 0 rows)
    role_map = {}
    if only_label0:
        zero_rows = df[df[label_col] == 0]
        rec_only = set(zero_rows[zero_rows[receptor_col].isin(only_label0)][receptor_col])
        lig_only = set(zero_rows[zero_rows[ligand_col].isin(only_label0)][ligand_col])
        for s in only_label0:
            r = s in rec_only
            l = s in lig_only
            role_map[s] = ("receptor+ligand" if (r and l) else ("receptor" if r else "ligand"))
    return only_label0, role_map

# Apply to judith_train
judith_label0_only_seqs, judith_label0_role_map = sequences_only_in_label0(judith_train)
print(f"Judith train: {len(judith_label0_only_seqs)} sequences occur only in label 0 pairs.")

# Example: inspect first 5
print("Sample (judith):")
for s in list(judith_label0_only_seqs):
    print(judith_label0_role_map[s], s)

# If a DataFrame is useful:
judith_label0_only_df = pd.DataFrame(
    [(s, judith_label0_role_map[s]) for s in judith_label0_only_seqs],
    columns=["sequence", "role_in_label0"]
)
```

```python
# Apply to judith_val
judith_label0_only_seqs, judith_label0_role_map = sequences_only_in_label0(judith_val)
print(f"Judith val: {len(judith_label0_only_seqs)} sequences occur only in label 0 pairs.")
```

```python
# Apply to judith_test
judith_label0_only_seqs, judith_label0_role_map = sequences_only_in_label0(judith_test)
print(f"Judith test: {len(judith_label0_only_seqs)} sequences occur only in label 0 pairs.")
```

# Create redundancy reduced subset of judiths sequences to test PINDER

```python
import pandas as pd
import os
```

```python
path = "/nfs/scratch/pinder/negative_dataset/my_repository/datasets/deleak_uniprot/deleak_cdhit"
train_df = pd.read_csv(f"{path}/train.csv")
val_df = pd.read_csv(f"{path}/val.csv")
test_df = pd.read_csv(f"{path}/test.csv")

# use only positives
train_df = train_df[train_df['label'] == 1]
val_df = val_df[val_df['label'] == 1]
test_df = test_df[test_df['label'] == 1]

# combine all
df = pd.concat([train_df, val_df, test_df], ignore_index=True)

# extract uniprot ids from 'entry' column
parts = df['entry'].astype(str).str.split('--', n=1, expand=True)
df['receptor_uniprot'] = parts[0].str.rsplit('_', n=1).str[-1]
df['ligand_uniprot'] = parts[1].str.rsplit('_', n=1).str[-1]

# unqiue uniprot ids
unique_receptor_uniprot = df['receptor_uniprot'].unique()
unique_ligand_uniprot = df['ligand_uniprot'].unique()
unique_uniprot = set(unique_receptor_uniprot).union(set(unique_ligand_uniprot))
print(f"Unique uniprot ids: {len(unique_uniprot)}")

# mapper uniprot ID to sequence
uniprot_to_seq = {}
for _, row in df.iterrows():
    uniprot_to_seq[row['receptor_uniprot']] = row['receptor_seq']
    uniprot_to_seq[row['ligand_uniprot']] = row['ligand_seq']

# save to file
path = path + "/unique_sequences_uniprot_id"
os.makedirs(path, exist_ok=True)
with open(f"{path}/unique_uniprot_ids.fasta", "w") as f:
    for uniprot_id in unique_uniprot:
        f.write(f">{uniprot_id}\n")
        f.write(f"{uniprot_to_seq[uniprot_id]}\n")
```

### remove identical pairs from my dataset


```python
my_path = "/nfs/scratch/pinder/negative_dataset/my_repository/datasets/deleak_uniprot/deleak_cdhit"
my_train = pd.read_csv(os.path.join(my_path, "train.csv"))
my_val = pd.read_csv(os.path.join(my_path, "val.csv"))
my_test = pd.read_csv(os.path.join(my_path, "test.csv"))

# keep only positives
my_train = my_train[my_train['label'] == 1]
my_val = my_val[my_val['label'] == 1]
my_test = my_test[my_test['label'] == 1]
```

```python
from collections import Counter
import random

def sample_negatives(df: pd.DataFrame, split: str, n_samples: int, path: bool = True) -> pd.DataFrame:
    """
    Sample negative pairs for a given split, matching the sequence frequency distribution of positives.

    Randomly pairs receptor and ligand sequences (not present as positives) with probability proportional
    to their frequency in the positive set.

    Args:
        df (pd.DataFrame): DataFrame of positive pairs for the split.
        split (str): Name of the split.
        n_samples (int): Number of negative samples to generate.

    Returns:
        pd.DataFrame: DataFrame of negative pairs.
    """

    positives = set(zip(df['receptor_seq'], df['ligand_seq']))
    rec_counter = Counter(df['receptor_seq'])
    rec_seqs    = list(rec_counter.keys())
    rec_weights = [rec_counter[s] for s in rec_seqs]
    lig_counter = Counter(df['ligand_seq'])
    lig_seqs    = list(lig_counter.keys())
    lig_weights = [lig_counter[s] for s in lig_seqs]

    neg_records = []
    while len(neg_records) < n_samples:
        rec_seq = random.choices(rec_seqs, weights=rec_weights, k=1)[0]
        lig_seq = random.choices(lig_seqs, weights=lig_weights, k=1)[0]
        
        if (rec_seq, lig_seq) in positives:
            continue

        if path:
            rec_path = df.loc[df['receptor_seq']==rec_seq, 'receptor_path'].iat[0]
            lig_path = df.loc[df['ligand_seq']==lig_seq,   'ligand_path'].iat[0]
            neg_records.append({
                'split': split,
                'receptor_seq': rec_seq,
                'ligand_seq':  lig_seq,
                'receptor_path': rec_path,
                'ligand_path':  lig_path,
                'label': 0
            })
        else:
            neg_records.append({
                'split': split,
                'receptor_seq': rec_seq,
                'ligand_seq':  lig_seq,
                'label': 0
            })

    return pd.DataFrame(neg_records)
```

```python
def remove_self_interactions(df, split: str):
    print(f"Original {split} dataset size: {len(df)}")
    df = df[df['receptor_seq'] != df['ligand_seq']]
    print(f"Filtered {split} dataset size: {len(df)}")
    return df

my_train = remove_self_interactions(my_train, "train")
my_val = remove_self_interactions(my_val, "val")
my_test = remove_self_interactions(my_test, "test")

my_train_neg = sample_negatives(my_train, "train", len(my_train), path=False)
my_val_neg = sample_negatives(my_val, "val", len(my_val), path=False)
my_test_neg = sample_negatives(my_test, "test", len(my_test), path=False)

my_train = pd.concat([my_train, my_train_neg], ignore_index=True).reset_index()
my_val = pd.concat([my_val, my_val_neg], ignore_index=True).reset_index()
my_test = pd.concat([my_test, my_test_neg], ignore_index=True).reset_index()
```

```python
# save dfs
out_path = "/nfs/scratch/pinder/negative_dataset/my_repository/datasets/deleak_uniprot/deleak_cdhit/no_self_interactions"

my_train.to_csv(os.path.join(out_path, "train.csv"), index=False)
my_val.to_csv(os.path.join(out_path, "val.csv"), index=False)
my_test.to_csv(os.path.join(out_path, "test.csv"), index=False)
```

### Unique sequences in self interactions ?

```python
import pandas as pd
import os
```

```python
my_path = "/nfs/scratch/pinder/negative_dataset/my_repository/datasets/deleak_uniprot/deleak_cdhit/unique_sequences_uniprot_id/deleak_cd_hit"
my_train = pd.read_csv(os.path.join(my_path, "train.csv"))
my_val = pd.read_csv(os.path.join(my_path, "val.csv"))
my_test = pd.read_csv(os.path.join(my_path, "test.csv"))

# keep only positives
my_train = my_train[my_train['label'] == 1]
my_val = my_val[my_val['label'] == 1]
my_test = my_test[my_test['label'] == 1]
```

```python
# unique sequences in train, val, test
train_seqs = set(my_train['receptor_seq']).union(set(my_train['ligand_seq']))
val_seqs = set(my_val['receptor_seq']).union(set(my_val['ligand_seq']))
test_seqs = set(my_test['receptor_seq']).union(set(my_test['ligand_seq']))

# sequences in pair where ligand_seq == receptor_seq
same_seq_train = my_train[my_train['receptor_seq'] == my_train['ligand_seq']]
same_seq_val = my_val[my_val['receptor_seq'] == my_val['ligand_seq']]
same_seq_test = my_test[my_test['receptor_seq'] == my_test['ligand_seq']]

train_seqs_same = set(same_seq_train['receptor_seq'])
val_seqs_same = set(same_seq_val['receptor_seq'])
test_seqs_same = set(same_seq_test['receptor_seq'])

print(f"Train unique sequences: {len(train_seqs)}, with same receptor and ligand: {len(train_seqs_same)}")
print(f"Left sequences: {len(train_seqs) - len(train_seqs_same)}")
print(f"Val unique sequences: {len(val_seqs)}, with same receptor and ligand: {len(val_seqs_same)}")
print(f"Left sequences: {len(val_seqs) - len(val_seqs_same)}")
print(f"Test unique sequences: {len(test_seqs)}, with same receptor and ligand: {len(test_seqs_same)}")
print(f"Left sequences: {len(test_seqs) - len(test_seqs_same)}")
```

```python
print(f"Train sequences self-interactions: {len(train_seqs_same)}, samples self-interactions: {len(same_seq_train)} - {len(train_seqs_same)} = {len(same_seq_train) - len(train_seqs_same)}")
print(f"Val sequences self-interactions: {len(val_seqs_same)}, samples self-interactions: {len(same_seq_val)} - {len(val_seqs_same)} = {len(same_seq_val) - len(val_seqs_same)}")
print(f"Test sequences self-interactions: {len(test_seqs_same)}, samples self-interactions: {len(same_seq_test)} - {len(test_seqs_same)} = {len(same_seq_test) - len(test_seqs_same)}")
```

```python
different_seq_train = my_train[my_train['receptor_seq'] != my_train['ligand_seq']]
different_seq_val = my_val[my_val['receptor_seq'] != my_val['ligand_seq']]
different_seq_test = my_test[my_test['receptor_seq'] != my_test['ligand_seq']]

print(f"Train samples with different receptor and ligand: {len(different_seq_train)}")
print(f"Val samples with different receptor and ligand: {len(different_seq_val)}")
print(f"Test samples with different receptor and ligand: {len(different_seq_test)}")
```

```python
len_train_seqs_not_self = len(train_seqs) - len(train_seqs_same)
len_val_seqs_not_self = len(val_seqs) - len(val_seqs_same)
len_test_seqs_not_self = len(test_seqs) - len(test_seqs_same)

# (How many samples are generated from self interactions) + (# of sequences not self-interacting)
train_possible_positives_self_interacting = (len(same_seq_train) - len(train_seqs_same)) + min(len_train_seqs_not_self, len(train_seqs_same)) + len(different_seq_train)
val_possible_positives_self_interacting = (len(same_seq_val) - len(val_seqs_same)) + min(len_val_seqs_not_self, len(val_seqs_same)) + len(different_seq_val)
test_possible_positives_self_interacting = (len(same_seq_test) - len(test_seqs_same)) + min(len_test_seqs_not_self, len(test_seqs_same)) + len(different_seq_test)

print("(len(same_seq_train) - len(train_seqs_same)) + min(len_train_seqs_not_self, len(train_seqs_same)) + len(different_seq_train)")
print(f"{len(same_seq_train)} - {len(train_seqs_same)} + min({len_train_seqs_not_self}, {len(train_seqs_same)}) + {len(different_seq_train)}")
print("val_possible_positives_self_interacting = (len(same_seq_val) - len(val_seqs_same)) + min(len_val_seqs_not_self, len(val_seqs_same)) + len(different_seq_val)")
print(f"{len(same_seq_val)} - {len(val_seqs_same)} + min({len_val_seqs_not_self}, {len(val_seqs_same)}) + {len(different_seq_val)}")
print("test_possible_positives_self_interacting = (len(same_seq_test) - len(test_seqs_same)) + min(len_test_seqs_not_self, len(test_seqs_same)) + len(different_seq_test)")
print(f"{len(same_seq_test)} - {len(test_seqs_same)} + min({len_test_seqs_not_self}, {len(test_seqs_same)}) + {len(different_seq_test)}\n")

print(f"Train positives possible: {train_possible_positives_self_interacting}, before: {len(my_train)}")
print(f"Val positives possible: {val_possible_positives_self_interacting}, before: {len(my_val)}")
print(f"Test positives possible: {test_possible_positives_self_interacting}, before: {len(my_test)}")
```
