# relevance plots

```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
receptor = np.load("/nfs/scratch/pinder/negative_dataset/my_repository/plots/lrp_receptor_mean.npy")
ligand = np.load("/nfs/scratch/pinder/negative_dataset/my_repository/plots/lrp_ligand_mean.npy")

receptor
```

```python
receptor[10]
```

```python
# cut receptor to check correctness
# receptor = receptor[:5000]
```

```python
AA_ALPHABET = "ACDEFGHIKLMNPQRSTVWYXBUZO"  # 20 characters
alphabet_size = len(AA_ALPHABET)
alphabet_size
```

```python
receptor.shape
```

```python
drop = set("BUZO") # non-standard amino acids
keep_idx = [i for i, aa in enumerate(AA_ALPHABET) if aa not in drop]
AA_ALPHABET_21 = ''.join(aa for aa in AA_ALPHABET if aa not in drop)

# receptor: flat -> (positions, 25) -> (25, positions)
receptor_grid = receptor.reshape(-1, len(AA_ALPHABET)).T
ligand_grid   = ligand.reshape(-1, len(AA_ALPHABET)).T

# assuming AA axis is axis 0 (your y-axis in imshow)
receptor_grid = receptor_grid[keep_idx, ...]
ligand_grid   = ligand_grid[keep_idx, ...]

receptor_grid.shape
```

```python
receptor_n_rows = alphabet_size
receptor_n_cols = receptor.shape[0] // receptor_n_rows
receptor_n_cols
```

```python
plt.figure(figsize=(30, 5))
plt.imshow(receptor_grid, aspect='auto', cmap='viridis')
plt.colorbar(label='Mean |R_in|')
plt.yticks(ticks=range(len(AA_ALPHABET_21)), labels=AA_ALPHABET_21)
plt.show()
```

```python
cut_receptor_grid = receptor_grid[:, :400]

plt.figure(figsize=(30, 5))
plt.imshow(cut_receptor_grid, aspect='auto', cmap='viridis')
plt.colorbar(label='Mean |R_in|')
plt.yticks(ticks=range(len(AA_ALPHABET_21)), labels=AA_ALPHABET_21)
plt.show()
```

```python
ligand_n_rows = alphabet_size
ligand_n_cols = ligand.shape[0] // ligand_n_rows
ligand_n_cols
```

```python
plt.figure(figsize=(30, 5))
plt.imshow(ligand_grid, aspect='auto', cmap='viridis')
plt.colorbar(label='Mean |R_in|')
plt.yticks(ticks=range(len(AA_ALPHABET_21)), labels=AA_ALPHABET_21)
plt.show()
```

```python
cut_ligand_grid = ligand_grid[:, :400]

plt.figure(figsize=(30, 5))
plt.imshow(cut_ligand_grid, aspect='auto', cmap='viridis')
plt.colorbar(label='Mean |R_in|')
plt.yticks(ticks=range(len(AA_ALPHABET_21)), labels=AA_ALPHABET_21)
plt.show()
```

```python
title_size = 30
label_size = 25
tick_size = 25

fig, axes = plt.subplots(4, 1, figsize=(30, 33))

plots = [
    (receptor_grid, 'Receptor Mean Relevance (full)'),
    (cut_receptor_grid, 'Receptor Mean Relevance (first 400 positions)'),
    (ligand_grid, 'Ligand Mean Relevance (full)'),
    (cut_ligand_grid, 'Ligand Mean Relevance (first 400 positions)'),
]

vmin = min(np.min(grid) for grid, _ in plots)
vmax = max(np.max(grid) for grid, _ in plots)

for ax, (grid, title) in zip(axes, plots):
    im = ax.imshow(grid, aspect='auto', cmap='viridis', vmin=vmin, vmax=vmax)
    ax.set_title(title, fontsize=title_size)
    ax.set_ylabel('AA', fontsize=label_size)
    ax.set_yticks(range(len(AA_ALPHABET_21)))
    ax.set_yticklabels(AA_ALPHABET_21, fontsize=tick_size)
    ax.tick_params(axis='x', labelsize=tick_size)
    ax.tick_params(axis='y', labelsize=tick_size)
    cbar = fig.colorbar(im, ax=ax, orientation='vertical', fraction=0.02, pad=0.02)
    cbar.ax.tick_params(labelsize=tick_size)

axes[-1].set_xlabel('Sequence Position', fontsize=label_size)
axes[-1].tick_params(axis='x', labelsize=tick_size)
plt.tight_layout()
plt.savefig("/nfs/scratch/pinder/negative_dataset/my_repository/plots/Relevance_color_plots.png")
```
